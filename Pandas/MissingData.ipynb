```python
import numpy as np
import pandas as pd
```

```python
#Finding Missing Data
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()  # Selection hota h to column wise hota h
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
# Removing Missing Number
df
```

```python
df.dropna()   # Drop hota h to row wise hota h
```

```python
df.dropna(thresh=3)
```

```python
# Filling the missing data 
df
```

```python
df.fillna(0)
```

```python
values = {'A':0,'B':100,'C':300,'D':400}
df.fillna(value=values)  #  Fill hota h to column wise fill hota h
```
