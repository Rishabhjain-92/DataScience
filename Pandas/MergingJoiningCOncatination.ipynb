```python
import numpy as np 
import pandas as pd 
```

```python
# Merging 2 data frame

employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on="employee_id",how="inner") # agr how nhi likhte to by default inner hi rheta h   Ye btata h ki jo employee id h uske inner commom element ko dekhna h
```

```python
pd.merge(employees,salaries,on="employee_id",how="outer")
```

```python
pd.merge(employees,salaries,on="employee_id",how="left")  # mtlb left wala mtlb employees k bais pr merge kr rha hu
```

```python
pd.merge(employees,salaries,on="employee_id",how="right")
```

```python
# Concatination of 2 data frame

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])   # Column k basis pr 
```

```python
pd.concat([df1,df2],axis=1)  #Rows k basis pr
```

**Joing 2 data frame**

```python

df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how='outer')
```

```python
df2.join(df1)
```
